## Terminal



### 1. ¿Qué es la terminal o consola?

Una terminal o consola es una de las formas para comunicarnos con un computador. Esta nos provee de una interfaz instrucciones o comandos al computador. Por ejemplo, ejecutar programas instalados previamente como Python y Jupyter Notebook, navegar por el sistema de archivos del computador, explorar qué archivos hay en determinadas carpetas, interactuar con git y github, entre otros. 

El funcionamiento de esta interfaz es a través de comando de línea de texto, es decir, su funcionamiento se basa en que el usuario escribe comandos junto con sus parámetros para que el sistema operativo entregue una respuesta. 

Para efectos de este curso, las dos principales ventajas de utilizar la terminal son:

1. Poder tener varias terminales abiertas y cada una utilizarla para diferentes propósitos. Por ejemplo, **ejecutar diferentes archivos Python de forma simultánea**.

2. Ser capaz de saber siempre desde qué carpeta estamos ejecutando un archivo. Lo cual nos reducirá la posibilidad de tener errores relacionados a **rutas relativas**.


### 2. ¿Cómo abrir la terminal?

* **Windows**: En la barra de inicio, puedes buscar "cmd" y abrir el programa ("Command prompt")
* **Linux**: Puedes usar la combinación de teclas ` Ctrl + Alt + T ` para abrir la terminal.
* **macOS**: En Finder, abre la carpeta /Aplicaciones/Utilidades y haz doble clic en "Terminal".


## Comandos

Los comandos corresponden al mecanismo donde el usuario se puede comunicar con el computador y solicitar que haga ciertas acciones. Para este curso, los comandos más importantes a conocer son 4: `pwd`, `dir` o `ls`, `cd` y `python`. 


### 1. Formato del comando

El formato para todo comando es el siguiente:

`comando argumentos* -opciones*`

* **`comando`**: corresponde al nombre del comando a llamar. Por ejemplo, `pwd`, `cd`, `python`, `pip`.

- **`argumentos*`**: corresponde a los parámetros que el comando va a utilizar para su ejecución. Dependiente del parámetro, estos argumentos pueden ser 0, 1, 2, etc. Algunos casos son:
    1. **`pwd`** es un comando que no necesita argumentos, solo escribimos dicho comando y funcionará.
    2. **`python`** es un comando que cambia su funcionamiento según la cantidad de argumentos. Si se ejecuta sin ningún argumento (`python`), se abrirá la consola de Python. Si se ejecuta con 1 argumento que corresponde a un archivo .py (`python ejemplo3.py`), se ejecutará el archivo correspondiente.

* **`-opciones`**: también conocidos como _flags_, es la forma de especificar opciones durante la ejecución del comando. Cada opción se escribe como `-nombre_de_la_opción parámetro*` en donde el parámetro puede ser opcional según la opción que estemos llamando. Veremos ejemplos de esto más adelante.

#### Windows

En Windows tenemos 2 comandos básicos: `cd` y `dir`

##### 1. Ver directorio actual (`cd`)

Del inglés **c**hange **d**irectory, este comando nos permite movernos en el sistema de archivos para acceder a diferentes directorios/carpetas. Cuando ejecutamos este comando sin ningún argumento, nos retorna la posición actual como una **ruta absoluta**.

```cmd
➜ cd
C:\Users\Pedro
```
En este caso, nuestra terminal está ubicada dentro de la carpeta `Pedro` que a su vez esta carpeta está dentro de la carpeta `Users`, la cual está dentro de la carpeta raíz.


##### 2. Cambiar directorio actual (`cd`)

En caso de que se agregue una ruta (absoluta o relativa) al comando, este nos permitirá cambiar nuestro directorio.

```cmd
➜ cd
C:\Users\Pedro
➜ cd Libros
➜ cd
\Users\Pedro\Libros
```
Ahora, desde la carpeta `Pedro` nos movimos a la carpeta `Libros`. También es posible subir a la carpeta superior utilizando el argumento `..`.

```cmd
➜ cd
C:\Users\Pedro
➜ cd Libros
➜ cd
C:\Users\Pedro\Libros
➜ cd ..
➜ cd
C:\Users\Pedro
➜ cd Música
➜ cd
C:\Users\Pedro\Música
➜ cd ..\..\data
➜ cd
\data
```


##### 3. Listar contenido de un directorio (`dir`)

Del inglés **dir**ectory, este comando nos entrega un detalle de los archivos y carpetas presentes en un determinado directorio. De forma informal, podemos decir que este comando nos responde a la pregunta: ¿Qué tenemos en la carpeta?

```cmd
➜ cd
\Users\Pedro
➜ dir
... (texto omitido para este ejemplo)

Directorio de \Users\Pedro
02-08-2022  12:29    <DIR>          .
02-08-2022  12:29    <DIR>          ..
19-07-2022  18:45    <DIR>          Fotos
19-07-2022  18:45    <DIR>          Libros
19-07-2022  18:45    <DIR>          Música
... (texto omitido para este ejemplo)
➜ cd Libros
➜ cd
\Users\Pedro\Libros
➜ dir
... (texto omitido para este ejemplo)

Directorio de \Users\Pedro\Libros
02-08-2022  12:29    <DIR>          .
02-08-2022  12:29    <DIR>          ..
02-08-2022  12:29            11.960 calculo.pdf
02-08-2022  11:55            16.119 python.pdf
... (texto omitido para este ejemplo)

```

#### Linux y macOS

Tanto Linux como MacOS X funcionan con facilidades estándar de Unix en sus terminales. Esto nos permite utilizar los mismos comandos para ambos sistemas operativos: `pwd`, `cd` y `ls`


##### 1. Ver directorio actual (`pwd`)

De las siglas en inglés **p**rint **w**orking **d**irectory, es el comando que nos retorna la **ruta absoluta** para llegar a nuestra posición desde la terminal. De forma informal, podemos decir que este comando nos responde a la pregunta: ¿Dónde estamos?. Cuando ejecutamos este comando, nos retorna la posición actual como una **ruta absoluta**.

```cmd
➜ pwd
/Users/Pedro
```
En este caso, nuestra terminal está ubicada dentro de la carpeta `Pedro` que a su vez esta carpeta está dentro de la carpeta `Users`, la cual está dentro de la carpeta raíz.


##### 2. Cambiar directorio actual (`cd`)

Del inglés **c**hange **d**irectory, este comando nos permite movernos en el sistema de archivos para acceder a diferentes directorios/carpetas. Para esto, es necesario agregar una **ruta absoluta** o **relativa** como argumento.

```cmd
➜ pwd
/Users/Pedro
➜ cd Libros
➜ pwd
/Users/Pedro/Libros
```
Ahora, desde la carpeta `Pedro` nos movimos a la carpeta `Libros`. También es posible salir de una carpeta utilizando el argumento `..`.

```cmd
➜ pwd
/Users/Pedro
➜ cd Libros
➜ pwd
/Users/Pedro/Libros
➜ cd ..
➜ pwd
/Users/Pedro
➜ cd Música
➜ pwd
/Users/Pedro/Música
➜ cd ../../data
➜ pwd
/data
```


##### 3. Listar contenido de un directorio (`ls`)

Del inglés **l**i**s**t, este comando nos entrega un detalle de los archivos y carpetas presentes en un determinado directorio. De forma informal, podemos decir que este comando nos responde a la pregunta: ¿Qué tenemos en la carpeta?

```cmd
➜ pwd
/Users/Pedro
➜ ls
Fotos
Libros
Música
➜ Libros
➜ cd
/Users/Pedro/Libros
➜ ls
calculo.pdf
python.pdf
```

## Git y Github

Git es un sistema distribuido de **control de versiones** ampliamente utilizado en el mundo del desarrollo de software. A continuación, podrás ver los comandos básicos que necesitarás para poder utilizar git correctamente a lo largo del curso.

## Entendiendo Git

Si bien Git provee muchas herramientas, en el curso solo será necesario entender lo básico para poder desenvolverse de manera correcta en el curso. Para eso, necesitamos entender qué es un repositorio en git, y cómo se maneja.

En términos prácticos, un repositorio es un almacenamiento virtual de un conjunto de archivos, y almacena no solo el estado final de los archivos que contiene, sino también almacena _todas las versiones_ que se han ido almacenando en él. Al utilizar git, manejaremos 4 ubicaciones de los archivos: la carpeta del computador, la lista de cambios, el repositorio local y el repositorio remoto.

## Entendiendo Github

Github es una plataforma basada en la nube para alojar repositorios Git. Al tener todos los repositorios personales (privados y públicos) en un mismo lugar, facilita la colaboración en proyectos Git de forma online, por lo que su principal uso es compartir código y colaborar con otros.

## Comandos básicos

Para utilizar git en el curso, solo necesitarás 5 comandos a lo largo de todo el semestre: `init`, `clone`, `pull`, `add`, `commit` y `push`. Todos los comandos de git puedes correrlos desde la consola, por lo que si tienes alguna duda por ese lado puedes volver a leer el material de como usar una terminal.

### git init
Este comando permitirá crear un directorio _oculto_ `.git` _en la carpeta donde queramos alojar un repositorio local_ (no lo verás si no tienes configurado tu computador para mostrar los archivos ocultos). Esta carpeta será utilizada por Git para almacenar toda la información necesaria para gestionar el historial de versiones de tu repositorio. Por lo mismo, si la eliminas, Git perderá el hisotrial del proyecto.

### git clone
Con este comando podrás clonar un repositorio ya existente en tu computador, como el repositorio de los contenidos que usaremos en este curso. Para esto, puedes abrir una terminal, ubicarte en la carpeta donde desees guardar el o los repositorios, y ejecutar, en este caso, el siguiente comando:

```
git clone https://github.com/jtmarquinezv/nivelacion_python_mia_202501
```
Si es que el clonado fue exitoso, entonces deberías ver en tu computador la carpeta con la Sesión 0 y Sesión 1.

### git pull
Como se mencionó anteriormente, está el repositorio remoto y el repositorio local. Si es que hay archivos o cambios que están en el repositorio remoto pero no están en el repositorio local de tu computador, para traerte los cambios debes hacer

```
git pull
```
Este comando lo necesitarás durante el curso para descargar el material nuevo para cada sesión y actualizar tu repositorio en caso de que utilices más de un computador para subir cambios.

### git add
Para entender el flujo de subida, vamos a ir ejemplificando con el siguiente escenario: Se tiene un archivo en el computador (archivo morado) llamado README.md, que uno quiere hacer que llegue al repositorio remoto. Para esto, tendremos que usar distintos comando de git para hacer que funcione. 
<center><img src="img/image.png" alt="Drawing" style="width: 800px;"/></center>
En una primera instancia, debemos informarle a git que tenemos la intención de agregar el archivo a nuestro repositorio. Para esto, podemos hacer

```
git add README.md
```
El comando recibe el nombre (o el path) del archivo que queremos agregar al repositorio. Una vez hecho esto, nuestro archivo pasa a estar en la lista de cambios, pero aún no está en nuestro repositorio. En la lista de cambios podemos poner tantos archivos como queramos, por lo que podemos ir agregando varios de una misma vez, para luego agregarlos al repositorio todos juntos (para ello, podríamos haber escrito `git add .`, en que ese `.` representaría "todos los archivos"). Una vez hecho el `git add README.md`, el estado quedaría así:

Importante notar, que todavía no se ha añadido el archivo ni los cambios al repositorio, por lo que aún no se ha generado una versión de nuestros archivos dentro del repositorio (local ni remoto).

### git commit
Como mencionamos anteriormente, git es un sistema de control de versiones. Para crear una versión de nuestros archivos, debemos usar la instrucción `commit`. Para crear una versión, podemos ejecutar el comando 

```
git commit -m "mensaje descriptivo explicando lo que estoy agregando al repositorio"
```
(Es con comilla doble. Con comilla simple puede generar problema) Una vez hecho esto, se crea una versión de nuestros archivos, y se agregan todos los cambios y archivos que estaban en la lista de cambios a una nueva versión dentro del repositorio. Estas versiones guardan una serie de datos (que muchas no son tan relevantes para el curso), pero queda un registro de la hora en la que se hizo el commit, y los cambios y archivos que se incluyeron con dicho commit. Habiendo hecho esta instrucción, nuestro estado ahora es el siguiente:
<center><img src="img/image-3.png" alt="Drawing" style="width: 800px;"/></center>
El archivo morado ya es parte del repositorio local, pero aún no está en el repositorio remoto.

### git push
La última instrucción que nos servirá para este curso, es el `push`. Lo que hace esto, es publicar y subir al repositorio remoto todas las versiones (o commits) que estén en el repositorio local que aún no hayan sido subidas al repositorio remoto. Sin embargo, es práctica correcta verificar si hay cambios en el repositorio remoto antes, para actualizar el código con las posibles versiones nuevas antes de "sobreescribir" el historial". Para hacer esto, basta con ejecutar 

```
git pull
git push
```
y se subirán los cambios al repositorio remoto, y el estado ahora es el siguiente:

## Trabajo colaborativo y resolución de conflictos

Supongamos que hay dos personas que están trabajando sobre el mismo archivo en su misma versión:

```
Línea 0
Línea 1
Línea 2
Línea 3
Línea 4
Línea 5
```

Supongamos que un desarrollador A está trabajando sobre este archivo en su repositorio local y agregó líneas de código entre las líneas 1 y 2; y que, por su parte, un desarrollador B (colaborador autorizado del mismo repositorio) está trabajando sobre este mismo archivo en su repositorio local y agrega líneas de código entre las líneas 4 y 5. Si ambos hacen sus commits correspondientes y luego hacen pull y push para que esos cambios se vean reflejados en el repositorio, git (y github) sabe que estos cambios son independientes y, por ende, sabe cómo unir los cambios, resultando en un archivo correcto:

```
Línea 0
Línea 1
Línea nueva de A
Línea nueva de A
Línea 2
Línea 3
Línea 4
Línea nueva de B
Línea nueva de B
Línea 5
```

Supongamos ahora que ambos deciden trabajar entre la línea 2 y 3, pero cada uno en sus computadores, por lo que ambos escriben líneas de código. Cuando ambos hagan add, commit y pull (previo a push), git se va a encontrar ante un **conflicto** de versiones. ¿Qué líneas deberían ir antes? ¿Qué líneas después? ¿Cuál es la versión que debería quedar como final? ¿La de A, la de B o una mezcla de ambos? Cuando un desarrollador haga git pull, Git indicará que primero se deben resolver esos conflictos, incluyendo ciertos indicadores en el código. Por lo mismo, alguno de los dos (coordinado previamente aunque generalmente el que hizo pull y se encontró con el conflicto) al abrir el código se encontrará con algo como lo siguiente:

```
Línea 0
Línea 1
Línea nueva de A
Línea nueva de A
Línea 2
<<<<<<< HEAD
Tus cambios
=======
Los cambios que alguien más hizo
>>>>>>> branch-name
Línea 3
Línea 4
Línea nueva de B
Línea nueva de B
Línea 5
```

Los indicadores (_conflict markers_) `<<<<<<<`, `=======` y `>>>>>>>` aparecerán siempre y el encargado de resolver el conflicto deberá revisar ambos códigos y (quizás con la ayuda de la otra persona) decidir si dejar la versión de A, la versión de B, o un arreglo distinto. Una vez que se tenga resuelto el conflicto en nuestro computador, eliminar los indicadores de conflicto y dejar sólo la versión final del código, guardar, hacer `git add .`, hacer `git commit -m 'mensaje que indica resolucion de conflicto'` (esto resuelve el conflicto), y finalmente hacer push de los cambios para que queden reflejados en el repositorio remoto.

Nota: Si git rechaza el push, es porque surgieron nuevos conflictos con nuevas modificaciones que surgieron en este proceso, y habría que repetir los pasos luego de hacer git pull para darse cuenta qué ocurrió.
